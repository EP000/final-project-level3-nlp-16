```python
dataset = pd.read_csv("/opt/ml/final-project-level3-nlp-16/data/data.tsv", sep="\t")
```

```python
dataset
```

```python
dataset = dataset.astype({"pid":"str"})
```

```python
dataset = dataset.drop(["img"], axis=1)
```

```python
dataset
```

```python
img_path = Path("/opt/ml/final-project-level3-nlp-16/data/imgs")

pid_to_img = {}
for category in img_path.iterdir():
    category_path = img_path / category
    for file in category_path.iterdir():
        pid_to_img[PurePath(file).stem] = str(file)
```

```python
img_df = pd.DataFrame(pid_to_img, index=["img"]).T
img_df = img_df.reset_index()
img_df = img_df.rename(columns={"index":"pid"})
```

```python
img_df
```

```python
dataset
```

```python
dataset = dataset[dataset["pid"].isin(list(pid_to_img.keys()))].reset_index(drop=True)
dataset
```

```python
dataset = pd.merge(dataset, img_df, on="pid")
```

```python
dataset
```

```python
dataset.to_csv("/opt/ml/final-project-level3-nlp-16/data/data.tsv", sep='\t', index=False)
```

```python
dict_label_to_num = {}
dict_num_to_label = {}
for idx, category in enumerate(sorted(dataset["label"].unique())):
    dict_label_to_num[category] = idx
    dict_num_to_label[idx] = category
```

```python
import pickle
with open("/opt/ml/data/final_project/code/dict_label_to_num", "wb") as f:
    pickle.dump(dict_label_to_num, f, protocol=pickle.HIGHEST_PROTOCOL)
with open("/opt/ml/data/final_project/code/dict_num_to_label", "wb") as f:
    pickle.dump(dict_num_to_label, f, protocol=pickle.HIGHEST_PROTOCOL)
```
