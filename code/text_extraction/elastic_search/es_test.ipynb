```python
import es_api as es
from pprint import pprint as pp

query = "갤럭시 S8 64기가 상태굿 ㅡ중고폰3977 갤럭시s8 64기가 오키드그레이 상태굿 최초 LG개통했으나 모든유심다됩니다 선불.알뜰유심다됩니다 화면에 사진처럼 잔상조금있어 싸게판매합니다 ㅡ사진참조 터치및 모든 기능이상없이 작동잘됩니다 직거래는 인천 1호선 주안역입니다 택배거래도 가능합니다 이0ㅡ29오오ㅡ99이사"
res_list = es.es_search(query)
res_list = res_list['hits']['hits']
```

```python
result = {}
er = []
for doc in res_list: #res['hits']['hits']
    for term in doc["_explanation"]["details"]:
        try:
            word = term["description"]
            word = word.split(' ')[0][13:]
            if word not in result:
                idf_value = term['details'][0]['details'][1]['value'] 
                result[word] = idf_value
        except IndexError as error:
            er.append(term)
```

```python
# "갤럭시 s8 64기가 상태굿 ㅡ중고폰3977 갤럭시s8 64기가 오키드그레이 상태굿 최초 LG개통했으나 모든유심다됩니다 선불.알뜰유심다됩니다 화면에 사진처럼 잔상조금있어 싸게판매합니다 
# ㅡ사진참조 터치및 모든 기능이상없이 작동잘됩니다 직거래는 인천 1호선 주안역입니다 택배거래도 가능합니다 이0ㅡ29오오ㅡ99이사"
result_k = list(result.keys())
query_token = query.split(' ')

tf_value = {}

for key in query_token:
    if key in result:
        if tf_value.get(key):
            tf_value[key] += 1
        else:
            tf_value.setdefault(key, 1)
        

print(tf_value)
```

```python
for term in result:
    result[term] = result[term] * tf_value[term]
```
